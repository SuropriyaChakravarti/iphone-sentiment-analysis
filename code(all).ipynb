## Perform sentiment analysis

### Subtask:
Calculate sentiment scores for each review and determine the sentiment category (positive, negative, or neutral).


**Reasoning**:
Import the necessary VADER sentiment analysis library, define functions to calculate sentiment score from tokens and categorize sentiment, and apply these functions to the DataFrame.



```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment_score(tokens):
    if isinstance(tokens, list):
        text = " ".join(tokens)
        return analyzer.polarity_scores(text)['compound']
    return 0.0

df['sentiment_score'] = df['review_body_lemmatized_tokens'].apply(get_sentiment_score)

def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['sentiment_score'].apply(categorize_sentiment)

display(df[['review_body_cleaned', 'sentiment_score', 'sentiment_category']].head())
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon', quiet=True)

analyzer = SentimentIntensityAnalyzer()

def get_sentiment_score(tokens):
    if isinstance(tokens, list):
        text = " ".join(tokens)
        return analyzer.polarity_scores(text)['compound']
    return 0.0

df['sentiment_score'] = df['review_body_lemmatized_tokens'].apply(get_sentiment_score)

def categorize_sentiment(score):
    if score >= 0.05:
        return 'Positive'
    elif score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment_category'] = df['sentiment_score'].apply(categorize_sentiment)

display(df[['review_title', 'review_body', 'sentiment_score', 'sentiment_category']].head())
```
